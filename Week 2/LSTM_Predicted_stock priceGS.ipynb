## Transform the Data

```python
X0=[]
X1=[]
X2=[]
X3=[]
X4=[]
X5=[]
y=[]
```

```python
for i in range (0, GS_data.shape[0]-48):
    X0.append(GS_data.iloc[i:i + 48, 0])
    X1.append(GS_data.iloc[i:i + 48, 1])
    X2.append(GS_data.iloc[i:i + 48, 2])
    X3.append(GS_data.iloc[i:i + 48, 3])
    X4.append(GS_data.iloc[i:i + 48, 4])
    X5.append(GS_data.iloc[i:i + 48, 5])
    y.append(GS_data.iloc[i + 48, 6])
```

```python
type(X5)
```

```python
X5
```

```python
GS_data.head(50)
```

```python
X0, X1, X2, X3, X4, X5, y = np.array(X0), np.array(X1), np.array(X2), np.array(X3), np.array(X4), np.array(X5),np.array(y)

```

```python
y.shape
```

```python
y=np.reshape(y, (len(y),1))
y.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0,1))
X0 = scaler.fit_transform(X0)
X1 = scaler.fit_transform(X1)
X2 = scaler.fit_transform(X2)
X3 = scaler.fit_transform(X3)
X4 = scaler.fit_transform(X4)
X5 = scaler.fit_transform(X5)
y = scaler.fit_transform(y)
```

```python
X5
```

```python
X5.shape
```

```python
X = np.stack([X0, X1, X2, X3, X4, X5], axis=2)
```

```python
X.shape
```

The results have 3 dimensions

the 1st number in the output is Samples. One sequence is one sample. A batch is comprised of one or more samples.

2nd number in the output is Time Steps. One time step is one point of observation in the sample.

The third number of the output is Feature. One feature is one observation at a time step.

## Build the LSTM Model

```python
X_train, X_test = X[:-480], X[-480:]
y_train, y_test = y[:-480], y[-480:]
```

```python
X_train.shape[1], X_train.shape[2]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
from tensorflow.keras import optimizers
```

```python
model = Sequential()
# set return requesnce = true, if we are going to have another LSTM layer
model.add(LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(LSTM(50, return_sequences=False))
model.add(Dense(50, activation='relu'))
model.add(Dense(1))
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping

filepath = 'models/{epoch:02d}-{loss:.4f}-{val_loss:.4f}-{mae:.4f}-{val_mae:.4f}.hdf5'

callbacks = [EarlyStopping(monitor='val_loss', patience=20),
             ModelCheckpoint(filepath, monitor='loss', save_best_only=True, mode='min')]
```

```python
optimizers.SGD(momentum=0.9)
model.compile(optimizer='SGD', loss='mse', metrics=['mae'])
```

```python
model.fit(X_train, y_train, validation_split=0.2, epochs=1000, callbacks=callbacks, batch_size=16)
```

```python
MSE, MAE = model.evaluate(X_test, y_test)
print(MSE, MAE)
```

```python
predictions = model.predict(X_test)
predictions = scaler.inverse_transform(predictions)
predictions
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-paper')
plt.figure(figsize=(24,12))
plt.title('GS stock price predicted')
plt.xlabel('Years', fontsize=14)
plt.ylabel('Stock price', fontsize=18)
plt.plot(predictions)
plt.plot(scaler.inverse_transform(y_test))
plt.legend(['Forecast', 'Actual'])
plt.show()
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(24,12))
plt.title('30 Minutes ahead Electricity Price Forecasting')
plt.xlabel('Step (30 Minutes)', fontsize=14)
plt.ylabel('Electricity Price (AUD)', fontsize=18)
plt.plot(predictions[200:800])
plt.plot(scaler.inverse_transform(y_test)[200:800])
plt.legend(['Forecast', 'Actual'])
plt.show()
```
