```r
library(rcompanion)
library(readxl)
library(ggplot2)
library(IntroCompFinR)
library(methods)
library(dplyr)
library(lubridate)
```

```r
MorganStanley = read.csv(r"(C:\Users\Mary\Documents\Data Science Classes\Final Project\Week 2\MS Historical Data.csv)")

MorganStanley
```

```r
# What is the average per day?

MorganStanley$rowMeans = rowMeans(MorganStanley[ , c(2,3,4,5)], na.rm=TRUE)

MorganStanley$rowMeans
```

```r
MorganStanley$Date <- gsub('/', '-', MorganStanley$Date)
MorganStanley
```

```r
MorganStanley$Date = as.Date(MorganStanley$Date, format= "%m-%d-%Y")
```

```r
MorganStanley
```

```r
MorganStanley$Day <- lubridate::day(MorganStanley$Date)
MorganStanley$Month <- lubridate::month(MorganStanley$Date)
MorganStanley$Year <- lubridate::year(MorganStanley$Date)
```

```r
MorganStanley
```

```r
options(repr.plot.width=15, repr.plot.height=8)

ggplot(MorganStanley, aes(x=Date, y=rowMeans, group = Day)) +
  geom_line(color="red") +
  ggtitle("Average Price per Day") +
  theme(axis.text.x = element_text(angle = 90, vjust = 1, hjust=1)) 
```

```r
options(repr.plot.width=15, repr.plot.height=8)

ggplot(MorganStanley, aes(x=Year, y=rowMeans, group = Month)) +
  geom_line(color="red") +
  ggtitle("Average Price per Month") +
  theme(axis.text.x = element_text(angle = 90, vjust = 1, hjust=2))
```

```r
options(repr.plot.width=15, repr.plot.height=8)

ggplot(MorganStanley, aes(x=Year, y=rowMeans, group = Month)) +
  geom_line(aes(color=Month)) +
  ggtitle("Average Price per Year By Month") +
  theme(axis.text.x = element_text(angle = 90, vjust = 1, hjust=2)) +
  theme(plot.title = element_text(size = 40, face = "bold", color="darkblue", hjust=0.5))
```

```r
options(repr.plot.width=15, repr.plot.height=8)

ggplot(MorganStanley, aes(x=Date, y=rowMeans, group = Year)) +
  geom_line(color="sky blue") +
  ggtitle("Average Price per Year") +
  theme(axis.text.x = element_text(angle = 90, vjust = 1, hjust=1))
```

```r
# Difference between High and Low Price by row from 1999 to 2022

highLowDiff = MorganStanley[, 4] - MorganStanley[, 5]

MorganStanley$highLowDiff = highLowDiff

MorganStanley$highLowDiff
```

```r
options(repr.plot.width=15, repr.plot.height=8)

ggplot(MorganStanley, aes(x=Date, y=highLowDiff, group = Year)) +
  geom_line(color="green") +
  ggtitle("Difference between High & Low Price per Year") +
  theme(axis.text.x = element_text(angle = 90, vjust = 1, hjust=1))
```

```r
# Have the prices gotten higher or lower from 1999-2022?

options(repr.plot.width=15, repr.plot.height=8)

ggplot(MorganStanley, aes(x=Date, y=Open, group = Year)) +
  geom_line(aes(color=Year)) +
  ggtitle("Have the Prices gotten Higher or Lower from 1999 to 2022") +
  theme(axis.text.x = element_text(angle = 90, vjust = 1, hjust=2)) +
  theme(plot.title = element_text(size = 30, face = "bold", color="dark blue", hjust=0.5))
```

```r
# Difference between Open & Closing Prices

OpenCloseDiff = MorganStanley[, 3] - MorganStanley[, 2]

MorganStanley$OpenCloseDiff = OpenCloseDiff

MorganStanley$OpenCloseDiff
```

```r
options(repr.plot.width=15, repr.plot.height=8)

ggplot(MorganStanley, aes(x=Date, y=OpenCloseDiff, group = Day)) +
  geom_line(aes(color=Month)) +
  ggtitle("Difference between Open & Close Prices By Month") +
  theme(axis.text.x = element_text(angle = 90, vjust = 1, hjust=2)) +
  theme(plot.title = element_text(size = 30, face = "bold", color="dark blue", hjust=0.5))
```

```r
# Does the # of Shares Bought (Volume) affect the Closing Price?

options(repr.plot.width=15, repr.plot.height=8)

ggplot(MorganStanley, aes(x=Price, y=Vol., group = Year)) +
  geom_line(color="orange") +
  ggtitle("Volume bought per Closing Price per Year") +
  theme(axis.text.x = element_text(angle = 90, vjust = 1, hjust=1)) 
```
