```r
library(rcompanion)
library(readxl)
library(ggplot2)
library(IntroCompFinR)
library(methods)
library(dplyr)
library(lubridate)
library(corrplot)
library(gmodels)
library(matrixStats)
library(scales)
```

```r
GoldmanSachs = read_xlsx(r"(C:\Users\Mary\Documents\Data Science Classes\Final Project\Excel Master Data Sets\GS Master.xlsx)")
MorganStanley = read_xlsx(r"(C:\Users\Mary\Documents\Data Science Classes\Final Project\Excel Master Data Sets\MS Master.xlsx)")
Schwab = read_xlsx(r"(C:\Users\Mary\Documents\Data Science Classes\Final Project\Excel Master Data Sets\SCHW Master.xlsx)")
Unemployment = read_xlsx(r"(C:\Users\Mary\Documents\Data Science Classes\Final Project\Week 3\Unemployment Rates 2000-2021.xlsx)")
```

```r
GoldmanSachs$Day <- lubridate::day(GoldmanSachs$Date)
GoldmanSachs$Month <- lubridate::month(GoldmanSachs$Date)
GoldmanSachs$Year <- lubridate::year(GoldmanSachs$Date)
MorganStanley$Day <- lubridate::day(MorganStanley$Date)
MorganStanley$Month <- lubridate::month(MorganStanley$Date)
MorganStanley$Year <- lubridate::year(MorganStanley$Date)
Schwab$Day <- lubridate::day(Schwab$Date)
Schwab$Month <- lubridate::month(Schwab$Date)
Schwab$Year <- lubridate::year(Schwab$Date)
```

```r
# Stock Price per Year for Goldman Sachs, Morgan Stanley, and Schwab

options(repr.plot.width=15, repr.plot.height=8)

ggplot() +
  geom_line(data = GoldmanSachs, aes(x = Date, y = Open, group = Year), color = "blue") + # must include argument label "data"
  geom_line(data = MorganStanley, aes(x = Date, y = Open, group = Year), color = "green") +
  geom_line(data = Schwab, aes(x = Date, y = Open, group = Year), color = "red") +
  ggtitle("Price per year") +
  theme(plot.title = element_text(size = 40, face = "bold", color="darkblue", hjust=0.5))
```

```r
GoldmanSachsYear = aggregate(Open ~ Year, data = GoldmanSachs, FUN = sum)
MorganStanleyYear = aggregate(Open ~ Year, data = MorganStanley, FUN = sum)
SchwabYear = aggregate(Open ~ Year, data = Schwab, FUN = sum)
```

```r
options(repr.plot.width=15, repr.plot.height=8)

ggplot() + 
  geom_line(data = GoldmanSachsYear, aes(x = Year, y = Open), color = "blue") +
  geom_line(data = MorganStanleyYear, aes(x = Year, y = Open), color = "green") +
  geom_line(data = SchwabYear, aes(x = Year, y = Open), color = "red") +
  ggtitle("Price by Year") +
  theme(plot.title = element_text(size = 40, face = "bold", color="darkblue", hjust=0.5))
```

```r
# How did the unemployment rate affect stocks

options(repr.plot.width=15, repr.plot.height=8)

ggplot(Unemployment, aes(x = Year, y = Unemployment.Rate)) + 
  geom_line(color = "purple") +
  ggtitle("Unemployment Rate from 2000-2021") +
  theme(plot.title = element_text(size = 40, face = "bold", color="darkblue", hjust=0.5))
```

```r
# The Unemployment Rate was one of many things that played a factor in the rise & fall of the stocks. As the unemployment 
# rate went up, the stock prices also increased. The Feds would lower interest rates which would in turn cause investors 
# to buy more stocks.
```

```r
# What effect did the Housing Market Downfall

options(repr.plot.width=15, repr.plot.height=8)

Goldman2_df <- GoldmanSachs %>%
  select(Date, Open) %>%
  mutate(Year = year(Date)) %>%
  filter(year(Date) > 2005 & year(Date) < 2010) %>%
  group_by(Year) %>%
  mutate(Company = "Goldman Sachs")

Morgan2_df <- MorganStanley %>%
  select(Date, Open) %>%
  mutate(Year = year(Date)) %>%
  filter(year(Date) > 2005 & year(Date) < 2010) %>%
  group_by(Year) %>%
  mutate(Company = "Morgan Stanley")

Schwab2_df <- Schwab %>%
  select(Date, Open) %>%
  mutate(Year = year(Date)) %>%
  filter(year(Date) > 2005 & year(Date) < 2010) %>%
  group_by(Year) %>%
  mutate(Company = "Schwab")

Grouped2.df <- bind_rows(Goldman2_df, Morgan2_df, Schwab2_df)

ggplot(Grouped2.df, aes(x = Date, y = Open, group = interaction(Company, Year), color = Company)) +
  geom_line() +
  labs(title = "Stock Open Prices of Goldman Sachs, Morgan Stanley and Schwab (2006-2009)",
       x = "Date", y = "Open Price") +
  theme(plot.title = element_text(size = 25, face = "bold", color="darkblue", hjust=0.5))
```

```r
# When the Housing Market Crashed in 2006, more demand for stocks led to increased prices in stocks. As the 
# Great Recession hit, the demand for stocks decreased therefore,the price decreased based off the demand of the stocks.
# Near the end of the Recession in 2008, the demand for stocks started going up again. The Stock prices continued to rise 
# through most of 2009, even after the Housing Market Crash ended July 2009.

# The Unemployment Rate also played a factor in the rise & fall of the stocks. As the unemployment rate went up, the stock 
# prices also increased. The Feds would lower interest rates which would in turn cause investors to buy more stocks.
```

```r
# Using group by to Create a Line Graph with 2008 from the Great Depression

options(repr.plot.width=15, repr.plot.height=8)

GoldmanSachs %>%
  select(Date, Open) %>%
  mutate(Year = year(Date)) %>%
  filter(year(Date) > 2007 & year(Date) < 2009) %>%
  group_by(Year) %>%
  ggplot(aes(x = Date, y = Open, group = Year)) +
  geom_line(color = "blue") +
  ggtitle("Price of Goldman Sachs in 2008") +
  theme(plot.title = element_text(size = 40, face = "bold", color="darkblue", hjust=0.5))

MorganStanley %>%
  select(Date, Open) %>%
  mutate(Year = year(Date)) %>%
  filter(year(Date) > 2007 & year(Date) < 2009) %>%
  group_by(Year) %>%
  ggplot(aes(x = Date, y = Open, group = Year)) +
  geom_line(color = "green") +
  ggtitle("Price of Morgan Stanley in 2008") +
  theme(plot.title = element_text(size = 40, face = "bold", color="darkblue", hjust=0.5))

Schwab %>%
  select(Date, Open) %>%
  mutate(Year = year(Date)) %>%
  filter(year(Date) > 2007 & year(Date) < 2009) %>%
  group_by(Year) %>%
  ggplot(aes(x = Date, y = Open, group = Year)) +
  geom_line(color = "red") +
  ggtitle("Price of Schwab in 2008") +
  theme(plot.title = element_text(size = 40, face = "bold", color="darkblue", hjust=0.5))
```
