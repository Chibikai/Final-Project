```r
library(rcompanion)
library(readxl)
library(ggplot2)
library(IntroCompFinR)
library(methods)
library(dplyr)
library(lubridate)
library(corrplot)
```

```r
Schwab = read.csv(r"(C:\Users\Mary\Documents\Data Science Classes\Final Project\Week 2\SCHW.csv)")

Schwab
```

```r
# What is the average per day?

Schwab$rowMeans = rowMeans(Schwab[ , c(2,3,4,5)], na.rm=TRUE)

Schwab$rowMeans
```

```r
Schwab$Day <- lubridate::day(Schwab$Date)
Schwab$Month <- lubridate::month(Schwab$Date)
Schwab$Year <- lubridate::year(Schwab$Date)
```

```r
options(repr.plot.width=15, repr.plot.height=8)

ggplot(Schwab, aes(x=Date, y=rowMeans, group = Day)) +
  geom_line(color="#69b3a2") +
  ggtitle("Average Price per Day") +
  theme(axis.text.x = element_text(angle = 90, vjust = 1, hjust=1)) 
```

```r
options(repr.plot.width=15, repr.plot.height=8)

ggplot(Schwab, aes(x=Year, y=rowMeans, group = Month)) +
  geom_line(color="red") +
  ggtitle("Average Price per Month") +
  theme(axis.text.x = element_text(angle = 90, vjust = 1, hjust=2))
```

```r
options(repr.plot.width=15, repr.plot.height=8)

ggplot(Schwab, aes(x=Year, y=rowMeans, group = Month)) +
  geom_line(aes(color=Month)) +
  ggtitle("Average Price per Year By Month") +
  theme(axis.text.x = element_text(angle = 90, vjust = 1, hjust=2)) +
  theme(plot.title = element_text(size = 40, face = "bold", color="darkblue", hjust=0.5))
```

```r
options(repr.plot.width=15, repr.plot.height=8)

ggplot(Schwab, aes(x=Date, y=rowMeans, group = Year)) +
  geom_line(color="sky blue") +
  ggtitle("Average Price per Year") +
  theme(axis.text.x = element_text(angle = 90, vjust = 1, hjust=1))
```

```r
# Difference between High and Low Price by row from 1999 to 2022

highLowDiff = Schwab[, 3] - Schwab[, 4]

Schwab$highLowDiff = highLowDiff

Schwab$highLowDiff
```

```r
options(repr.plot.width=15, repr.plot.height=8)

ggplot(Schwab, aes(x=Date, y=highLowDiff, group = Year)) +
  geom_line(color="green") +
  ggtitle("Difference between High & Low Price per Year") +
  theme(axis.text.x = element_text(angle = 90, vjust = 1, hjust=1))
```

```r
# What is the Difference between Adjusted Closing Price and Regular Closing Price

ClosingDiff = Schwab[, 5] - Schwab[, 6]

Schwab$ClosingDiff = ClosingDiff

Schwab$ClosingDiff
```

```r
options(repr.plot.width=15, repr.plot.height=8)

ggplot(Schwab, aes(x=Date, y=ClosingDiff, group = Year)) +
  geom_line(color="gold") +
  ggtitle("Difference between Adjusted Close & Close per Year") +
  theme(axis.text.x = element_text(angle = 90, vjust = 1, hjust=1))
```

```r
# Have the prices gotten higher or lower from 1999-2022?

options(repr.plot.width=15, repr.plot.height=8)

ggplot(Schwab, aes(x=Date, y=Open, group = Year)) +
  geom_line(aes(color=Year)) +
  ggtitle("Have the Prices gotten Higher or Lower from 1999 to 2022") +
  theme(axis.text.x = element_text(angle = 90, vjust = 1, hjust=2)) +
  theme(plot.title = element_text(size = 30, face = "bold", color="dark blue", hjust=0.5))
```

```r
# Difference between Open & Closing Prices

OpenCloseDiff = Schwab[, 2] - Schwab[, 5]

Schwab$OpenCloseDiff = OpenCloseDiff

Schwab$OpenCloseDiff
```

```r
options(repr.plot.width=15, repr.plot.height=8)

ggplot(Schwab, aes(x=Date, y=OpenCloseDiff, group = Day)) +
  geom_line(aes(color=Month)) +
  ggtitle("Difference between Open & Close Prices By Month") +
  theme(axis.text.x = element_text(angle = 90, vjust = 1, hjust=2)) +
  theme(plot.title = element_text(size = 30, face = "bold", color="dark blue", hjust=0.5))
```

```r
# Is there a Significance between Adjusted Close Price & Close Price?

options(repr.plot.width=15, repr.plot.height=8)

ggplot(Schwab, aes(x=Close, y=Adj.Close, group = Year)) +
  geom_line(color="orange") +
  ggtitle("Significance between Close & Adjusted Close Price per Year") +
  theme(axis.text.x = element_text(angle = 90, vjust = 1, hjust=1))
```

```r
# Does the # of Shares Bought (Volume) affect the Closing Price?

options(repr.plot.width=15, repr.plot.height=8)

ggplot(Schwab, aes(x=Close, y=Volume, group = Year)) +
  geom_line(color="orange") +
  ggtitle("Volume bought per Closing Price per Year") +
  theme(axis.text.x = element_text(angle = 90, vjust = 1, hjust=1)) 
```
